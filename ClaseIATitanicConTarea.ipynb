```python
def bar_char(feature):
  survived = df[df["Survived"]==1][feature].value_counts()
  dead = df[df["Survived"]==0][feature].value_counts()
  test = pd.DataFrame([survived,dead])
  test.index = ["Survived","Dead"]
  return test.plot(kind="bar",stacked=True,figsize=(10,5))
```

##2 Creando nuevas vectore features:


```python
def extrar_Titulo(data):
  titles = data["Name"].str.extract("([A-Za-z]+)\.",expand = True)
  return titles
```

```python
df["Title"]=extrar_Titulo(df)
df["Title"].value_counts()
```

##Vamos a crear un mapa de titulo
Se tomaran los valores mas representativos
1. Mr:0
2. Miss:1
3. Mrs:2
4. Otros:3

```python
def replace_titles(data):
  title_mapping={"Mr":"señor","Miss":"señora","Mrs":"señorita"}
  data["Title"]=data["Title"].map(title_mapping).fillna("Otros")
  return data
```

```python
df=replace_titles(df)
df.sample(10)
```

```python
bar_char('Title')
```

##Eliminando features que no nos sirven

```python
df.drop(["Name","Ticket","Cabin"], axis=1, inplace=True)
df.head()
```

```python
title_age_means=df.groupby("Title")["Age"].mean()
title_age_means
```

```python
df["Age"]=df.apply(lambda row: title_age_means[row["Title"]] if pd.isnull(row["Age"]) else row["Age"],axis=1)
```

##Actividad
encontrar una estrategia que me clasifique en varios grupos las personas que subieron al titanic (Niños,Adolecentes,jovenadulto, adulto,adulto mayo)
rango de edades:Niños(0-12),adolecentes(13-19),joven adulto(20-35),36-60),adulto mayor(>60)

```python
def classify_age_group(age):
    if age <= 12:
        return 'Niño'
    elif 13 <= age <= 19:
        return 'Adolescente'
    elif 20 <= age <= 35:
        return 'Joven adulto'
    elif 36 <= age <= 60:
        return 'Adulto'
    else:
        return 'Adulto mayor'

df['Age_Group'] = df['Age'].apply(classify_age_group)
display(df.head())
```

```python
bar_char('Age_Group')
```

Basado en el gráfico de barras, podemos observar la distribución de supervivencia y fallecimientos entre los diferentes grupos de edad. Parece que el grupo de 'Joven adulto' es el que tiene el mayor número tanto de sobrevivientes como de víctimas. El grupo de 'Niño' parece tener una tasa de supervivencia relativamente más alta en comparación con otros grupos. El grupo de 'Adulto mayor' es el que tiene menos individuos en general, con un número bajo de sobrevivientes.
