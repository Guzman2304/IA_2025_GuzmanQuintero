```python
# Tratamiento de datos
import numpy as np
import pandas as pd

# Gr√°ficos
import matplotlib.pyplot as plt

# Preprocesado y modelado
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
from sklearn.tree import export_graphviz
from sklearn.tree import export_text
from sklearn.model_selection import GridSearchCV
from sklearn import metrics
from sklearn.metrics import mean_squared_error, r2_score
```

```python
datos=datos[["TV","Radio","Newspaper","Web","Sales"]]
datos.describe()
```

Datos de entrenamiento y datos de validacion


Se emplearan la semila 1279(random_state=2022) los datos se destribuiran de la siguiente manera


1.   70% oara entrenamiento
2.   30% para calidacion del modelo



```python
X_entrena,X_valida,Y_entrena,Y_calida=train_test_split(datos.drop(columns="Sales"),datos["Sales"],train_size=.7,random_state=1279)
```

```python
X_entrena.shape
```

##4.1 MODELANDO EL ARBOL DE DECISIONES

La clase de decisionTreeRegressor del modulo sklearn.tree tiene los siguientes parametros:

* Max_depth: profundidad mazima que puede alcanzar el arvol.
* min_samples_split: numero minimo de observaciones que debe detener el nodo para que pueda dividirse. Si el calor decimal se interpreta como la draccion total de observaciones de entrenamiento ceil
(min_samples_spilth*n samples)
* min_samples_√±eaf: numero minimo de observaciones que deben tener cada uno de los nodos hijos para que se produzca una division

```python
modelo_ar=DecisionTreeRegressor(max_depth=3,random_state=2022)
```

Usamos los datos de entrenamiento

```python
modelo_ar.fit(X_entrena,Y_entrena)
```

## VISUALIZACION DEL MODELO

```python
print(f"Profundidad del arbol:{modelo_ar.get_depth()}")
print(f"Numero de nodos terminales:{modelo_ar.get_n_leaves()}")
```

```python
plt.figure(figsize=(20, 10))
plot_tree(modelo_ar, feature_names=X_entrena.columns, filled=True, rounded=True)
plt.show()
```

```python
text_modelo=export_text(decision_tree=modelo_ar,feature_names=datos.drop(columns="Sales").columns)
print(text_modelo)
```

## PREDICCIONES
La importancia de cada rpedictor en el modelo se calcula como la reduccion total(normalizada) en el criterio de division. en el caso, el mse qie se consifue el predictor enlas divisiones en las que participa, si un predictor no se ha seleccionado ninfina division no se incluye en el modelado y us importancia es 0

```python
importancia_predictores=pd.DataFrame({"predictor":datos.drop(columns="Sales").columns,"importancia":modelo_ar.feature_importances_})
```

```python
importancia_predictores.sort_values(by="importancia",ascending=False)
```
