```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler #Normalizacion
from sklearn.metrics import confusion_matrix
```

```python
#Cargar datos
df = pd.read_excel("/datos-logisticos.xlsx")
df.head()
```

```python
X1="Productos-Lote"
X2="Tiempo-Entrega"
Y="Defectuoso"
```

```python
X=df[[X1,X2]].values
Y=df[Y].values
#Entrenar el modelo con los datos
model = LogisticRegression(solver="liblinear")
model.fit(X,Y)
```

```python
#Crear la maya con todos los datos
X_min, X_max = df[X1].min() - 0.5, df[X1].max() + 0.5
Y_min, Y_max = df[X2].min() - 0.5, df[X2].max() + 0.5
h=(X_max - X_min)/100
xx,yy=np.meshgrid(np.arange(X_min,X_max,h),np.arange(Y_min,Y_max,h))
```

```python
#pREDICE EN TODOS LOS PPUNTOS DE LA MALLA
Z=model.predict(np.c_[xx.ravel(),yy.ravel()])
Z=Z.reshape(xx.shape)
```

```python
#Crear registros de grafico de dispersion
plt.figure(figsize=(10,6))
plt.scatter(df[Y==0][X1],df[Y==0][X2], color="Blue", label="No defectuoso")
plt.scatter(df[Y==1][X1],df[Y==1][X2], color="Red", label="Defectuoso")
#Dibuja la frontera de decision
plt.contourf(xx,yy,Z,alpha=0.4)
#Etiqueta de leyenda
plt.xlabel(X1)
plt.ylabel(X2)
plt.legend()
plt.show()
```
