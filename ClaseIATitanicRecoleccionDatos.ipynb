##2 Creando nuevas vectore features:


```python
def extrar_Titulo(data):
  titles = data["Name"].str.extract("([A-Za-z]+)\.",expand = True)
  return titles
```

```python
df["Title"]=extrar_Titulo(df)
df["Title"].value_counts()
```

##Vamos a crear un mapa de titulo
Se tomaran los valores mas representativos
1. Mr:0
2. Miss:1
3. Mrs:2
4. Otros:3

```python
def replace_titles(data):
  title_mapping={"Mr":"seor","Miss":"seora","Mrs":"seorita"}
  data["Title"]=data["Title"].map(title_mapping).fillna("Otros")
  return data
```

```python
df=replace_titles(df)
df.sample(10)
```

```python
bar_char('Title')
```

```python
df.columns
```

##Eliminando features que no nos sirven

```python
df.drop(["Name","Ticket","Cabin"], axis=1, inplace=True)
df.head()
```

```python
title_age_means=df.groupby("Title")["Age"].mean()
title_age_means
```

```python
df["Age"]=df.apply(lambda row: title_age_means[row["Title"]] if pd.isnull(row["Age"]) else row["Age"],axis=1)
```

##Actividad
encontrar una estrategia que me clasifique en varios grupos las personas que subieron al titanic (Nios,Adolecentes,jovenadulto, adulto,adulto mayo)
rango de edades:Nios(0-12),adolecentes(13-19),joven adulto(20-35),36-60),adulto mayor(>60)

```python
def classify_age_group(age):
    if age <= 12:
        return 'Nio'
    elif 13 <= age <= 19:
        return 'Adolescente'
    elif 20 <= age <= 35:
        return 'Joven adulto'
    elif 36 <= age <= 60:
        return 'Adulto'
    else:
        return 'Adulto mayor'

df['Age_Group'] = df['Age'].apply(classify_age_group)
display(df.head())
```

```python
bar_char('Age_Group')
```

Basado en el grfico de barras, podemos observar la distribucin de supervivencia y fallecimientos entre los diferentes grupos de edad. Parece que el grupo de 'Joven adulto' es el que tiene el mayor nmero tanto de sobrevivientes como de vctimas. El grupo de 'Nio' parece tener una tasa de supervivencia relativamente ms alta en comparacin con otros grupos. El grupo de 'Adulto mayor' es el que tiene menos individuos en general, con un nmero bajo de sobrevivientes.

###4.2 BINNING
Convertir una variable numerica a categorica


Mapeo de feature vector
* child: 0
* young: 1
* adult: 2
* mid_age: 3
* senior: 4

```python
for dataset in df:
  dataset.loc[dataset["Age"]<=12,"Age"]=0,
  dataset.loc[(dataset["Age"]>12) & (dataset["Age"]<=18),"Age"]=1,
  dataset.loc[(dataset["Age"]>18) & (dataset["Age"]<=22),"Age"]=2,
  dataset.loc[(dataset["Age"]>22) & (dataset["Age"]<=27),"Age"]=3,
  dataset.loc[(dataset["Age"]>2)]
```

```python
# Define the age bins and labels
bins = [0, 12, 19, 35, 60, np.inf]
labels = ['child', 'young', 'adult', 'mid_age', 'senior']

# Create a new column 'Age_Group_Binned' using pd.cut
df['Age_Group_Binned'] = pd.cut(df['Age'], bins=bins, labels=labels, right=True)

# Display the value counts for the new 'Age_Group_Binned' column
display(df['Age_Group_Binned'].value_counts())

# You can also display the first few rows of the dataframe to see the new column
display(df.head())
```
