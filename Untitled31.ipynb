##CALCULAR GINI y RMSE.

```python
from sklearn.metrics import mean_squared_error, roc_auc_score
import numpy as np

# Calcular GINI para el modelo TITANIC
# Gini is calculated for classification tasks. For decision trees,
# the gini impurity is a measure of the impurity of the split.
# We can calculate the gini coefficient of the predictions.
# However, a more common metric for classification is accuracy or AUC.
# If you meant Gini impurity at each node, that is part of the tree structure itself
# and not a single metric for the whole model performance like RMSE for regression.
# Assuming you want to calculate the Gini coefficient of the predictions vs actuals:
# This requires the actual and predicted values. Let's use the test set.

# Make predictions on the test set for Titanic
y_pred_titanic = titanic_tree.predict(X_test_titanic)

# Calculate Gini coefficient. Note: Gini coefficient is typically used
# in economics to measure income inequality. In machine learning,
# Gini impurity is used in decision trees, and Gini coefficient
# can be related to the AUC of the ROC curve (Gini = 2*AUC - 1).
# Let's calculate the Gini related to AUC.


try:
    auc_titanic = roc_auc_score(y_test_titanic, y_pred_titanic)
    gini_titanic = 2 * auc_titanic - 1
    print(f"Gini Coefficient (related to AUC) for TITANIC: {gini_titanic:.4f}")
except ValueError:
    print("Cannot calculate AUC/Gini for TITANIC. Need more than one class in test data.")
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Calcular Exactitud (Accuracy) para el modelo TITANIC
accuracy_titanic = accuracy_score(y_test_titanic, y_pred_titanic)
print(f"Exactitud (Accuracy) para el dataset TITANIC: {accuracy_titanic:.4f}")

# Mostrar Reporte de Clasificación para el modelo TITANIC
print("\nReporte de Clasificación para el dataset TITANIC:")
print(classification_report(y_test_titanic, y_pred_titanic, target_names=['No Superviviente', 'Superviviente']))
```
