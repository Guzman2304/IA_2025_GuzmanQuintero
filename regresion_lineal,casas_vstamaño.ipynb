```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
data = pd.read_excel("/content/Salary.xlsx")
data.head()
```

```python
df=pd.DataFrame(data) #Crear un dataframe
df.head()
```

```python
df.columns
```

```python
X=df[['YearsExperience']] #variable independiente
y=df['Salary'] #vareiable dependiente

plt.figure(figsize=[10,6])
plt.scatter(X,y, color="green",label="Datos Reales")
plt.title("Salario vs Años de Experiencia")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Ajustando los datos a un modelo de regresion lineal.
model = LinearRegression()
model.fit(X,y)

#Generar predicciones
y_pred = model.predict(X)

#Hallar a R2
r2=r2_score(y,y_pred)
print("R2: ",r2)
```

```python
plt.figure(figsize=[10,6])
plt.scatter(X,y, color="green",label="Datos Reales")
plt.plot(X,y_pred,color="red",linewidth=2,label="Regresion Lineal")
plt.title("Salario vs Años de Experiencia")
plt.xlabel("Años de Experiencia")
plt.ylabel("Salario")
plt.legend()
plt.grid(True)
plt.show()

print(f"Ecuacion de la recta: y= {model.coef_[0]:.2f}x+ {model.intercept_:2f}")
print(f"Coefficiente de Determinacion (R^2):{r2:.2f}")
```

```python
data2 = pd.read_csv("/content/dataset_edad_precio_seguro.csv")
data2.head()
```

```python
X=data2[['Edad']] #variable independiente
y=data2['PrecioSeguro'] #vareiable dependiente

plt.figure(figsize=[10,6])
plt.scatter(X,y, color="green",label="Datos Reales")
plt.title("PrecioSeguro vs Edad")
plt.xlabel("Edad")
plt.ylabel("PrecioSeguro")
plt.legend()
plt.grid(True)
plt.show()
```
